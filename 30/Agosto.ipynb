2.   Define ahora una función tal que, dado un conjunto de números enteros, regrese un subconjunto del conjunto tal que la suma de los elementos del subconjunto sea 0. Si no existe tal conjunto, la función debe regresar False.



```python
from random import randint
from itertools import combinations
def Numeros(Cantidad,R):
    Lista = {randint(-R, R) for i in range(Cantidad)}
    return Lista

Numeros(9,10)



```

```python
def todos_subconjuntos(conjunto):
    todos = []
    for i in range(len(conjunto)+1):
        for subconjunto in combinations(conjunto, i):
            todos.append(set(subconjunto))
    return todos

todos_subconjuntos(Numeros(9,10))
```
